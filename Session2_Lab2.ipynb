Building Visualizations to Understand the Dataframe and its Features

```python
plt.bar(housing_df['bedrooms'], housing_df['price'])
plt.xlabel('Bedrooms')
plt.ylabel('Price In Million')
plt.title('Bedrooms Vs Price In Million')
plt.show()
```

```python
plt.scatter(housing_df['bedrooms'], housing_df['price'])
plt.xlabel('Bedrooms')
plt.ylabel('Price In Million')
plt.title('Bedrooms Vs Price In Million')
plt.show()
```

```python
housing_df.plot(x='price',y='area',kind='line')
plt.xlabel('Price')
plt.ylabel('Area')
plt.title('Price Vs Area')
plt.show()
```
